##### 회의실 배정 (Greedy Algorithm)

백준 1931

- 한 개의 회의실, 사용하고자 하는 N개의 회의
- 각 회의에 대해 시작시간과 끝나는 시간 주어지며, 겹치지 않게 하는 회의 최대 개수 찾기
- 회의 시작시간과 끝나는 시간 같을 수 있음

- 첫 줄 회의 수 N (1 <= N <= 100000)
- 둘째 줄부터 N+1 줄까지 시작시간과 끝나는 시간

최대 사용할 수 있는 회의 최대 개수 출력

회의 시간을 짧으나 회의 시간이 긴 두 시간 사이에 딱 들어가 있을때 잡지 못하는 오류

```python
N = int(input())
T = []
for _ in range(N):
    s, e = map(int, input().split())
    dur = e - s
    T.append((s, e, dur))

T.sort(key=lambda x:x[-1])
print(T)

schedule = [0] * (max(max(T))+1)
schedule[0] = 1
count = 0

for x in T:
    if schedule[x[0]:x[1]] == [0]*(x[1]-x[0]):
        count += 1
        schedule[x[0]:x[1]] = [1]*(x[1]-x[0])
        print(schedule)
    else:
        continue

print(count)
```

- 회의실을 사용하고 있을때 가장 많은 회의가 열리기 위해서는 빨리 끝나야 한다 -> 종료시간이 빨라야 함
- 즉, 시작 시간을 기준으로 정렬하고 그 상태로 끝나는 시간이 빠른 순으로 정렬을 다시하여 꼬리잡기

```python
n = int(input())
room = []

for i in range(n):
    a, b = map(int, input().split())
    room.append([a, b])

room.sort(key = lambda x: x[0])
room.sort(key = lambda x: x[1])
# print(room)

cnt = 1
end = room[0][1]
for i in range(1, n):
    if room[i][0] >= end:
        cnt += 1
        end = room[i][1]
        # print(end)

print(cnt)
```

##### 보물 (Greedy Algorithm)

백준 1026

- 길이가 N인 정수 배열 A와 B가 있다.
- S = A[0]B[0] + A[1]B[1] + ... + A[N-1]B[N-1]
- S의 값을 가장 작게 만들기 위해 A의 수를 재배열하자

- 첫째 줄에 N (0 < N < 51 자연수)
- 둘째 줄에 A에 있는 N개의 수가 순서대로 주어지고
- 셋째 줄에 B에 있는 N개의 수가 순서대로 주어진다


```python
N = int(input())
A = list(map(int, input().split()))
A.sort(reverse=False)
B = list(map(int, input().split()))
B.sort(reverse=True)
C = list(zip(A,B))

result = 0
for x in C:
    result += x[0]*x[1]
print(result)
```
